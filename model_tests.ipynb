```python
import numpy as np
from sklearn.ensemble import ExtraTreesClassifier,AdaBoostClassifier,RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.externals import joblib
from sklearn.model_selection import cross_val_score

from data_utils import load_train, load_test, write_test
```

```python
X, y = load_train('./data/train_2008.csv')
```

```python
param_grid = {
    'n_estimators': [20, 50, 100, 500, 1000],
    'criterion':  ['gini', 'entropy'],
    'max_features': [None, 'auto', 'sqrt', 'log2'],
    'max_depth': [2, 5, 10, 20]
}

etc = ExtraTreesClassifier()
clf = GridSearchCV(estimator=etc, scoring='accuracy', param_grid=param_grid, 
                   n_jobs=4, verbose=2)
clf.fit(X, y)

print('Best Error: %f' %(clf.best_score_))
print('Best Model: %s' %(clf.best_params_))
```

```python
clf = ExtraTreesClassifier(n_estimators = 1000, criterion = 'entropy', max_depth = 20)
clf.fit(X,y)
```

```python
X_test = load_test('./data/test_2008.csv')
```

```python
vals = clf.predict(X_test)
```

```python
write_test("extratrees_preds.csv",vals)
```

```python
joblib.dump(clf,"extratrees1000n.pkl")
```

```python
param_grid = {
    'n_estimators': [20, 50, 100, 500],
    'algorithm':[ 'SAMME', 'SAMME.R'],
    'learning_rate': np.logspace(-4, 0, 10)
}

abc = AdaBoostClassifier()
clf = GridSearchCV(estimator=abc, scoring='accuracy', param_grid=param_grid, 
                   n_jobs=2, verbose=2)
clf.fit(X, y)

print('Best Error: %f' %(clf.best_score_))
print('Best Model: %s' %(clf.best_params_))
```
